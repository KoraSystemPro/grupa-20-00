```python
import numpy as np

a = np.array([1, 3, 6, 10])
print('Niz a:')
print(a)
print('oblik =', a.shape)
print('dimenzija =', a.ndim)
print('tip =', a.dtype)

M = np.array([[1, 2, 3], [4, 5, 6]], dtype='float32')
print(M)
print('oblik =', M.shape)
print('dimenzija =', M.ndim)
print('tip =', M.dtype)
```

```python
# Pravi niz sa odgovarajucim korakom
a1 = np.arange(10)
print('a1 =', a1)
a2 = np.arange(25, 30)
print('a2 =', a2)
a3 = np.arange(10, 50, 5)
print('a3 =', a3)

# Ekvidistantne tacke
a4 = np.linspace(2, 3, 5)
print('a4 =', a4)

# Niz koji se sastoji samo od nula
a5 = np.zeros(7)
print('a5 =', a5)

# Generisanje nausmicnih brojeva:
a6 = np.random.rand(3) # (n,m) puta random izmedju 0 i 1
print('a6 =', a6)
a7 = np.random.uniform(2, 5, 3) # izmedju 2-5, 3 nasumicna broja
print('a7 =', a7)

# Matrice
M1 = np.ones((2, 3))
print('M1 =\n', M1)
M2 = np.eye(4)
print('M2 =\n', M2)
M3 = np.diag([1, 2, 3, 4])
print('M3 =\n', M3)
M4 = np.random.rand(4, 3)
print('M4 = \n', M4)

# Operacije sa matricama
M1 = np.array([[1, 3, 5], [7, 9, 11], [13, 15, 17]])
M2 = np.array([[2, 4, 6], [8, 10, 12], [14, 16, 18]])
print('M1 + M2 =\n', M1 + M2)
print('M1 * M2 =\n', M1 * M2)
print('M1 x M2 =\n', M1.dot(M2))
print('M2 x M1 =\n', M2.dot(M1))
print('sin(M1) =\n', np.sin(M1))
```

```python
import matplotlib.pyplot as plt

x = np.arange(0, 2*2*np.pi, 1)
print(x)
y = np.sin(x)
print(y)

p = np.arange(0, 2*2*np.pi, 0.1)
q = np.sin(p)

plt.scatter(x, y, color='red')
plt.plot(p, q, color='green')
plt.show()
```

```python
# CSV
# Ime,Prezime,Razred,Odeljenje
# Marko,Petrovic,7,2
# Marija,Milenkovic,8,4
# Petar,Leposavic,7,5
import pandas as pd

data = pd.read_csv("advertising.csv", index_col=0)

print(data)
```

```python
sales = data['Sales']
tv = data['TV']
radio = data['Radio']
novine = data['Newspaper']
```

```python
plt.xlabel("TV")
plt.ylabel("Prodaje")
plt.scatter(tv, sales)
plt.show()
```

```python
plt.xlabel("Radio")
plt.ylabel("Prodaje")
plt.scatter(radio, sales, color="green")
plt.show()
```

```python
plt.xlabel("Novine")
plt.ylabel("Prodaje")
plt.scatter(novine, sales, color="red")
plt.show()
```

# Linearna regresija

## Prodaje = x * TV + y * Radio + z * Novine

```python
# TV
plt.xlabel("TV")
plt.ylabel("Prodaje")
plt.scatter(tv, sales)

fit, p = np.polyfit(tv, sales, 1)

plt.plot(tv, fit*tv + p, color="red")
plt.show()
print("x:", fit)
```

```python
# Radio
plt.xlabel("Radio")
plt.ylabel("Prodaje")
plt.scatter(radio, sales)

fit, p = np.polyfit(radio, sales, 1)

plt.plot(radio, fit*radio + p, color="red")
plt.show()
print("y:", fit)
```

```python
# Novine
plt.xlabel("Novine")
plt.ylabel("Prodaje")
plt.scatter(novine, sales)

fit, p = np.polyfit(novine, sales, 1)

plt.plot(novine, fit*novine + p, color="red")
plt.show()
print("z:", fit)
```
